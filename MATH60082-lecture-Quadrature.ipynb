# Lecture 10: Quadrature

## Department of Mathematics


Consider the well-known Black and Scholes (1973) partial differential equation for an option with an underlying
asset following geometric Brownian motion:
$$
\frac{\partial{V}}{\partial{t}} + \frac{1}{2} {\sigma}^2 S^2
\frac{\partial {^2V}}{\partial{S^2}} + (r-D_c) S
\frac{\partial{V}}{\partial{S}} - rV = 0,
$$
where $V(S,t)$ is the price of the derivative product, $S$ the
current value of the underlying asset, $t$ is time,  $T$ is the time to maturity
 $r$ the risk-free interest rate, $\sigma$ the
volatility of the underlying asset and $X$ is the exercise price of the option. 
$D_c$ is a continuous dividend yield which could, for example, be the foreign 
interest rate in a foreign exchange option.

Next make the following standard
transformations                
$$
x = \log(S_0),
$$
$$
y = \log(S_{T}).
$$
The final conditions for a European option expiring at time $T$
with  $V(y,T)$ are transformed in straightforward fashion; e.g. 
the payoff for a call option $V(S,T)  = \max(S_{T}-X,0)$ becomes
$$
\quad V(y,T) = \max(e^y-X,0). 
$$
The value of 
the option at time $t=0$ on an underlying asset  $S_0$  
has an  exact solution given by 
$$
V(x,0) = A (x) \int_{-\infty}^{\infty} B(x,y)V(y,T) \, dy,
$$
where, 
$$
A(x) = \frac{1}{\sqrt{2\sigma^2\pi T}} e^{-\frac12 kx-\frac18 
\sigma^2k^2 T - r T},
$$
and
$$
B(x,y) = e^{-\frac{(x-y)^2}{2\sigma^2 T} + \frac12 ky}
$$
and
$$
k = \frac{2(r-D_c)}{{\sigma}^2} -1.
$$

Let us go through an example solving this for a simple call option, before moving onto a barrier option. We will need to declare the functions $A$, $B$, a method for integrating and a method to return the value $V(x,0)$. Since integrations must work on uniformly-spaced grids, we also provide a function to do this. First include libraries:

```python
import numpy as np
import scipy
import matplotlib.pyplot as plt

from math import exp, log, sqrt
```

As we will be integrating numerically we also will need finite upper and lower bounds for the region of integration, so we need a way to input these values. In fact the integration will look like
$$
V(x,0) = A (x) \int_\text{yMin}^\text{yMax} B(x,y)V(y,T) \, dy \approx A(x) \sum_j w_j B(x,y_j)V(y_j,T),
$$
where the weights $w_j$ depend on which numerical integration scheme we choose.
By using this approximation we are assuming that the payoff is zero for $y<\text{yMin}$ and $y>\text{yMax}$, so in some cases the approximation can be improved here by using analytic results for the integrals either side
$$
V_\text{min}(x,0) = A (x) \int_{-\infty}^\text{yMin} B(x,y)V(y,T) \, dy,
$$
and
$$
V_\text{max}(x,0) = A (x) \int_\text{yMax}^\infty B(x,y)V(y,T) \, dy,
$$
such that
$$
V(x,0) = V_\text{min} + A (x) \int_\text{yMin}^\text{yMax} B(x,y)V(y,T) \, dy + V_\text{max},
$$

Define a function that evaluates $A$ as given by the formula:

```python
# the function A
def A_func( x, r, sigma, k, dt):
    return 1. / sqrt(2 * sigma*sigma*np.pi*dt) * exp(-0.5*k*x - 0.125*sigma*sigma*k*k*dt - r*dt)
```

Define a function that evaluates $B$ :

```python
# the function B
def B_func( x,  y,  sigma,  k,  dt):
    return exp(-(x - y)*(x - y) / (2.*sigma*sigma*dt) + 0.5 * k * y)
```

# Example - Call Option

We first set up the grid over the range of initial prices by applying a log transform
$$
x = \log(S_0)
$$
and for the $y$ variable at $T$ for a call option we use
$$
y \in [ \log(X) , \log (X) + \xi \sigma \sqrt{T} ] ,
$$
where $\xi$ is some grid scale factor set large enough to capture the value of the option.
After the parameter $k$ has been evaluated, we can generate values of the integrand, $f_j$
$$
f_j = B(x,y_j) V(y_j,T) = B(x,y_j) (e^{y_j} -X)
$$
for our call option.

Then, using a numerical scheme called [Simpson's rule](https://en.wikipedia.org/wiki/Simpson's_rule), the value of the option can be found as
$$
V(x,0) = A(x) \frac{\text{yMax}-\text{yMin}}{3 n}\left( f_0 + 2\sum_{j=1}^{n/2} f_{2j} + 4\sum_{j=0}^{n/2} f_{2j+1}+ f_n \right)
$$

Now setup some model parameters:

```python
S0 = 97.7402
X = 100.
r = 0.06
Dc = 0.02
sigma = 0.2
T = 1.
```

We need to setup the grid $y$ at the terminal time $t=T$:

```python
N=500
xi=10.0
    
# now get value of x at time t=0
x = log(S0)

# and setup the vector y
y = np.linspace(log(X), log(X)+ 10*sigma*sqrt(T),N+1)
```

Now enter the values for $f_j$ as defined above

```python
# Parameter k
kStar = 2.*(r - Dc) / sigma / sigma - 1.
           
# Set up a vector to store values of the integrand.
# It must be the same size as the y vector
f = np.zeros(N+1)
    
for j in range(N+1):
    # note here that $y_j>\log(X)$ so the option is always exercised
    f[j] = B_func( x , y[j] , sigma , kStar , T ) * (exp(y[j]) - X)
    
# Now use SciPy to do a Simpson's rule integration
# We begin by computing the fixed spacing between points in the y grid.
h = y[1] - y[0]
value = A_func(x, r, sigma, kStar, T )*scipy.integrate.simpson(y=f,dx=y[1]-y[0])
print("C(",S0,") = ", value) 
```

Bundle the calculation into a function so that we can investigate easily the convergence.

```python
def  callOption( S0, X, r, Dc, sigma, T, N, xi):
    # now get value of x at time t=0
    x = log(S0)
    # and setup the vector y
    y = np.linspace(log(X),log(X)+ 10*sigma*sqrt(T),N+1)
        
    # variable k
    kStar = 2.*(r - Dc) / sigma / sigma - 1.
            
    # setup vectors to store option values, and integrand function
    # these vectors must be the same size as the y vector
    f = np.zeros(N+1)
        
    for j in range(N+1):
        # note here that $y_j>\log(X)$ so the option is always exercised
        f[j] = B_func( x , y[j] , sigma , kStar , T ) * (exp(y[j]) - X)
        
    # now use scipy to do a Simpsons rule integration
    # first get the fixed space inbetween points in y
    h = y[1] - y[0]
    # then intergrate function f
    return A_func(x, r, sigma, kStar, T )*scipy.integrate.simpson(y=f,dx=y[1]-y[0])
    
```

Construct a convergence table to show what happens as $N\to\infty$.

```python
S0 = 97.7402
X = 100.
r = 0.06
Dc = 0.02
sigma = 0.2
T = 1.
   
# Examine convergence of a call option using QUAD...
print("|         N|  V(S,0;N)|diff ratio| conv rate|")
print("|----------|----------|----------|----------|")

valueOld = 1.
diffOld=1.

nRatio = 2

for k in range(0,8):
    n = int( 10* (nRatio**k) )
    value = callOption(S0,X,r,Dc,sigma,T,n,10.)
    diff = value - valueOld
    diffRatio = diffOld/diff
    if k<2:
        convrate=np.nan
    else:
        convrate=log(diffRatio)/log(nRatio)
    print(f"|{n:10d}|{value:10.3f}|{diffRatio:10.3f}|{convrate:10.3f}|")
    valueOld = value
    diffOld = diff
    
```

Now we can see the power of this method, since we use integration with Simpson's rule, we are able to get 4th order convergence,
$$
V^N = V^\text{exact} + O\left(N^{-4}\right) ,
$$
and we could get even higher convergence by extrapolating these results!

## Example &mdash; Up and Out Barrier Call Option

This is nice, but we already have an analytic solution to this problem. How about something more complicated? The discretely observed, up-and-out barrier call option is a problem that can't easily be solved with standard techniques such as Monte-Carlo or PDE methods. However this problem is uniquely adaptable to quadrature methods, given that  we know at each observation time point $t_k$ the value at which the payoff becomes zero, the region over which this happens can be split out and set to zero.

To see how to solve this problem using quadrature, the best way is to first observe how the grid adapts to the problem through time. Consider an up and out barrier call option observed at equally spaced times $t_k=k\Delta t$ such that $$\Delta t = \frac{T}{K}.$$ The payoff for the option is given by
$$
V(S,T) = \max(S-X,0)
$$
and the condition at each observation point is that
$$
V(S,t_k) = 0 \text{ if } S>B \text{ for some } k=0,1,\dots K-1 .
$$
As above, we apply a log transform at terminal time, so that the $y$ variable at $T$ becomes
$$
y \in [ \log(X) ,  \log(X) + \xi \sigma \sqrt{T} ] .
$$

Now for the first $x$ grid. This is for values of the optiopn at $t=t_{K-1}$, and we use the fact that $V$ is zero above the barrier to ignore any point above $S>B$, therefore
$$
x \in [  \log(S_0) - \xi \sigma \sqrt{T} , \log(B) ] .
$$

The lower boundary here is calculated to be sufficiently small such that $B( \log(S_0), y_\text{min} , t_k )V(y_\text{min},t_k) = 0 $ for $y<y_\text{min}$.

The remaining grids (those at $t = k\Delta t$ for $k \in \{1, 2, \dots K-2\}$ follow a similar pattern until we arrive at the last step, at which point we will wish to calculate $x_0 =  \log(S_0)$.

```python
S0 = 97.7402
X = 100.
r = 0.06
Dc = 0.02
sigma = 0.2
T = 1.

# K is the number of observations
K=5

# B is the barrier
B=110.

# N is the number of nodes at each step
N=4

# grid scale
xi=1.
    
# first setup a time grid
t = np.linspace(0.0,T,num=K+1)
```

Now set up the $y$ grid at $t=t_K=T$, and the $x$ grid at $t=t_{K-1}$

```python
y = np.linspace(log(X), log(X)+ xi*sigma*sqrt(t[K]),num=N+1)
x = np.linspace(log(S0)-xi*sigma*sqrt(t[K-1]), log(B),num=N+1)
```

Let's plot the way that values at $t=t_{K-1}$ depend on those at $t=t_K$. In the figure below the different colours correspond to the valuation carried out at a given node in the $x$ grid.

```python
# colours
colours = ['b','g','r','c','m','y','k']
plt.xlabel("t_k")
plt.ylabel("S")
# plot the grid nodes on the final step
for i in range(N+1):
    nodeColour = colours[ i%7 ]
    for j in range(N+1):
        plot_line_x = [ t[K-1] ,t[K] ]
        plot_line_y = [ exp(x[i]) , exp(y[j]) ]
        plt.plot(plot_line_x,plot_line_y,color = nodeColour)
```

And now, using a similar approach, we move backwards through time:

```python
# colours
colours = ['b','g','r','c','m','y','k']
plt.xlabel("t_k")
plt.ylabel("S")

# setup grid nodes at final step
y = np.linspace(log(X),log(X)+ xi*sigma*sqrt(t[K]),num=N+1)
x = np.linspace(log(S0)-xi*sigma*sqrt(t[K-1]),log(B),num=N+1)
# plot the grid nodes on the final step
for i in range(N+1):
    nodeColour = colours[ i%7 ]
    for j in range(N+1):
        plot_line_x = [ t[K-1] ,t[K] ]
        plot_line_y = [ exp(x[i]) , exp(y[j]) ]
        plt.plot(plot_line_x,plot_line_y,color = nodeColour)
        
# to move backwards through grid, copy x values onto the "y" grid,
y = np.copy(x)

# now move back through time
for k in range(K-2,0,-1):
    # and the x grid at t=t_{k}, now using the barrier as the upper limit and S0 at the lower limit
    x = np.linspace(log(S0)-xi*sigma*sqrt(t[k]),log(B),num=N+1)

    # plot the grid nodes on the final step
    for i in range(N+1):
        nodeColour = colours[ i%7 ]
        for j in range(N+1):
            plot_line_x = [ t[k] ,t[k+1] ]
            plot_line_y = [ exp(x[i]) , exp(y[j]) ]
            plt.plot(plot_line_x,plot_line_y,color = nodeColour)
    # overwrite the y values with x
    y = np.copy(x)

# and at the final step, x = x0 at t=t_{0}, 
x = np.linspace(log(S0),log(S0),num=1)

for i in range( len(x) ):
    nodeColour = colours[ i%7 ]
    for j in range(N+1):
        plot_line_x = [ t[0] ,t[1] ]
        plot_line_y = [ exp(x[i]) , exp(y[j]) ]
        plt.plot(plot_line_x,plot_line_y,color = nodeColour)
```

To work all the way back to $t=0$, we need only introduce a pair of arrays `vNew` and `vOld` and, steping back through time as we would in a PDE solver, value the option at time $t_k$ by approximating integrals over the values at $t_{k+1}$.

```python
# Black Scholes parameters
# K is the number of observations
# Barrier is the barrier
# N is the number of nodes at each step
# grid scale
def barrierCallOption(S0,X,r,Dc,sigma,T ,K,B,N,xi):
    # Set up a time grid
    t = np.linspace(0.0,T,num=K+1)
        
    # Set up option value grids for the final step.
    y = np.linspace(log(X), log(X)+ xi*sigma*sqrt(t[K]), num=N+1)
    x = np.linspace(log(S0)-xi*sigma*sqrt(t[K-1]), log(B), num=N+1)
    h = y[1] - y[0] # grid spacing for the integral
 
    # Allocate storage for the option values
    vNew = np.zeros(N+1)
    vOld = np.zeros(N+1)
    
    for j in range(N+1):
        vOld[j] = exp(y[j]) - X

    # Parameter k
    kStar = (2.*(r-Dc) / (sigma*sigma)) - 1.;

    # Set up a vector to store values of the integrand.
    # It must be the same size as the y vector.
    f = np.zeros(N+1)

    # Calculate the option values at x_i
    for i in range(N+1):            
        for j in range(N+1):
            # Note here that $y_j>\log(X)$ so the option is always exercised
            f[j] = B_func( x[i], y[j], sigma, kStar, T - t[K-1] ) * vOld[j]
          
        # Now use SciPy to do a Simpson's rule integration.
        vNew[i] = A_func(x[i], r, sigma, kStar, T - t[K-1] )*scipy.integrate.simpson(y=f,dx=h)       
    
    # Replace the y grid with the current x grid and replace the values in vOld with those in vNew.
    y = np.copy(x);
    vOld = np.copy(vNew);
    
    # Step back through time until we reach t=0.
    for k in range(K-2,0,-1):
        # Set up the x grid at t=t_{k}, now using the barrier as the upper limit and S0 at the lower limit
        x = np.linspace(log(S0)-xi*sigma*sqrt(t[k]), log(B), num=N+1)
        
        # Calculate value at every grid node at t[k]
        for i in range(N+1):
            for j in range(N+1):
                # note here that $y_j>\log(X)$ so the option is always exercised
                f[j] = B_func( x[i], y[j], sigma, kStar, t[k+1]-t[k] ) * vOld[j]
            
            # Use Simpson's rule
            h = y[1] - y[0]
            vNew[i] = A_func(x[i], r, sigma, kStar, t[k+1]-t[k] )*scipy.integrate.simpson(y=f, dx=h)       
        
        # Move on to the next step, replacing vOld with vNew and the y-gris with the x one.
        y = np.copy(x)
        vOld = np.copy(vNew)
    
    # At the final step, consider only x = x0 =log(S_0) at t=t_{0}, 
    for j in range(N+1):
        f[j] = B_func( log(S0), y[j], sigma, kStar, t[1]-t[0] ) * vOld[j]
        
    # Do one last integration with Simpson's rule
    return A_func(log(S0), r, sigma, kStar, t[1]-t[0] )*scipy.integrate.simpson(y=f, dx=y[1]-y[0])
    
```

So now we can examine the convergence for the up-and-out Barrier call option using QUAD:

```python
# BS parameters
S0 = 97.7402
X = 100.
r = 0.06
Dc = 0.02
sigma = 0.2
T = 1.

# number of observations
K=5

# B is the barrier
B=110.
    
   
# Examine convergence of a call option using QUAD...
print("|         N|  V(S,0;N)|diff ratio| conv rate|")
print("|----------|----------|----------|----------|")

valueOld = 1.
diffOld=1.

nRatio = 2

for k in range(0,8):
    n = int( 10* (nRatio**k) )
    value = barrierCallOption(S0,X,r,Dc,sigma,T,K,B,n,10.)
    diff = value - valueOld
    diffRatio = diffOld/diff
    if k<3:
        convrate=np.nan
    else:
        convrate=log(diffRatio)/log(nRatio)
        
    print(f"|{n:10d}|{value:10.6f}|{diffRatio:10.3f}|{convrate:10.3f}|")
    valueOld = value
    diffOld = diff
```

Again using integration and the simpsons rule, we are able to get 4th order convergence,
$$
V^N = V^\text{exact} + O\left(N^{-4}\right) .
$$
This is because we avoid any errors that arise from discontinuities in the payoff by splitting the integral. The method is tricky to setup and harder to generalise than other methods, but it is extremely powerful none the less.
